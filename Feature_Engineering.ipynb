Encoding of Categorical Data

Feature transformation: 
1. feature scaling : using Normalization and standardization
2. encoding categorical data:  ordinal and Nominal

```python
# In ordinal there is a defined order like Excellent, good, bad
# we do it using Ordinal encoding
# and one hot encoding is used for Nominal data

#Label encoding: this is used when our output labels i.e. the output class is in category
# for this we use label encoder of Sklearn
```

```python
# Ordinal encoder: workes when the categorcal data has some order
# ex: we have student qualificats data like post graduate (PG), UG, high school (HS)
# so while doing ordinal encoding sklearn will give numerical values as below for ethat column
# PG: 1
# UG: 2
# HS:3
```

```python
import pandas as pd
df = pd.read_csv('customer.csv')
```

```python
df.head()
```

```python
df1 = df.iloc[:,2:]
```

```python
df1.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(df1.iloc[:,0:2],df1.iloc[:,-1],test_size=0.2)
```

```python
from sklearn.preprocessing import LabelEncoder,OrdinalEncoder
```

```python
oe = OrdinalEncoder(categories=[["Poor","Average","Good"],["School","UG","PG"]])
oe.fit(X_train)
```

```python
oe.transform(X_train)
oe.transform(X_test)
```

```python
oe.categories_
```

```python
le = LabelEncoder()
le.fit_transform(y_train)
le.transform(y_test)
```

```python
le.classes_
```

#One Hot Encoding

used for nominal categorical data encoding

```python
# colour::::   red  green  blue
# blue          0    0      1
# green         0    1      0
# red           1     0      0

# like this we encode the categories of a column. this will add new coloumns
```

```python
# #Basically we are creating the dummy variables by onehot encoding.
# once the variables are created  we remove one  column from the dummy vaariables ( usually first column) to avoid the 
# dummy variable trap i.e from n columns to n-1 columns
# Dummy Variable trap: If we don't remove 1 dummy column then it will result in multicolinearity of our input variables which we
# don't want especially for our Linear models like logistic regression and linear regressions
```
